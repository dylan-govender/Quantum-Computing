```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its Taylor series expansion
def f(x):
    return np.exp(x)

def taylor_series_approximation(x, x0, max_n):
    approximation = 0
    for n in range(max_n + 1):
        term = (np.exp(x0) / math.factorial(n)) * ((x - x0) ** n)
        approximation += term
    return approximation

# Define the Maclaurin series approximation function for e^x
def maclaurin_series_approximation(x, max_n):
    approximation = 0
    for n in range(max_n + 1):
        term = (x ** n) / math.factorial(n)
        approximation += term
    return approximation

# Define the range of x values
x_values = np.linspace(-1, 2, 400)

# Point of expansion
x0 = 0.5

# Plot the Taylor series approximations
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Different truncation points for the series
n_values = [1, 2, 3, 5, 10]

# Create subplots for side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Define the range of x values for plotting
x_values = np.linspace(-1, 2, 400)

# True function for reference
true_values = f(x_values)

# Plot Maclaurin series
for i, n in enumerate(n_values):
    y_values_maclaurin = maclaurin_series_approximation(x_values, n)
    ax1.plot(x_values, y_values_maclaurin, label=f'Maclaurin (n={n})', color=colors[i])

# Plot Taylor series
x0 = 0.5  # Point of expansion for Taylor series
for i, n in enumerate(n_values):
    y_values_taylor = taylor_series_approximation(x_values, x0, n)
    ax2.plot(x_values, y_values_taylor, label=f'Taylor (n={n})', color=colors[i])

# Plot true function
ax1.plot(x_values, true_values, label='e^x (True Function)', color='black')
ax2.plot(x_values, true_values, label='e^x (True Function)', color='black')

# Set titles and labels for each subplot
ax1.set_title('Maclaurin Series Approximation of $e^x$')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.legend()
ax1.grid(True)

ax2.set_title('Taylor Series Approximation of $e^x$ at $x_0 = 0.5$')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
