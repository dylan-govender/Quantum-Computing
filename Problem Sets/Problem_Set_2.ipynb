```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the state coefficients
alpha = np.sqrt(2/3)
beta = np.exp(1j * np.pi/2) / np.sqrt(3)

# Calculate the components of the Bloch vector
r_x = 2 * np.real(np.conj(alpha) * beta)
r_y = 2 * np.imag(np.conj(alpha) * beta)
r_z = np.abs(alpha)**2 - np.abs(beta)**2

# Create a Bloch sphere
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Bloch sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='c', alpha=0.1)

# Plot the Bloch vector
ax.quiver(0, 0, 0, r_x, r_y, r_z, color='r', arrow_length_ratio=0.3)

# Set labels and aspect
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_aspect('auto')

plt.show()

# Output the Bloch vector for reference
print("Bloch Vector:", (r_x, r_y, r_z))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the state coefficients
alpha = np.sqrt(2/3)
beta = np.exp(1j * np.pi/2) / np.sqrt(3)

# Calculate the components of the Bloch vector
r_x = 2 * np.real(np.conj(alpha) * beta)
r_y = 2 * np.imag(np.conj(alpha) * beta)
r_z = np.abs(alpha)**2 - np.abs(beta)**2

# Calculate spherical coordinates (psi, theta) for the Bloch vector
theta = np.arccos(r_z / np.sqrt(r_x**2 + r_y**2 + r_z**2))
psi = np.arctan2(r_y, r_x)

# Create a Bloch sphere
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Bloch sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='c', alpha=0.1)

# Plot the Bloch vector
ax.quiver(0, 0, 0, r_x, r_y, r_z, color='r', arrow_length_ratio=0.3)

# Set labels and aspect
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_aspect('auto')

plt.show()

# Output the Bloch vector and spherical coordinates for reference
print("Bloch Vector:", (r_x, r_y, r_z))
print("Spherical Coordinates:")
print("Theta:", np.degrees(theta))  # Convert to degrees for easier interpretation
print("Psi:", np.degrees(psi))  # Convert to degrees for easier interpretation
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the quantum state |
alpha = np.sqrt(2/3)
beta = np.exp(1j * np.pi/2) / np.sqrt(3)
psi = np.array([alpha, beta])

# Pauli matrices
sigma_x = np.array([[0, 1], [1, 0]])
sigma_y = np.array([[0, -1j], [1j, 0]])
sigma_z = np.array([[1, 0], [0, -1]])

# Calculate the Bloch vector components
r_x = np.vdot(psi, sigma_x @ psi).real
r_y = np.vdot(psi, sigma_y @ psi).real
r_z = np.vdot(psi, sigma_z @ psi).real

# Bloch vector
r = np.array([r_x, r_y, r_z])

# Plot the Bloch vector on the Bloch sphere
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, r[0], r[1], r[2], color='r', arrow_length_ratio=0.1)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Plot the unit sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='b', alpha=0.1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

r
```
